# (ロジスティック)シグモイド関数

$\begin{eqnarray}
    sigmoid(z)=\frac{1}{1+e^{-z}} \nonumber
\end{eqnarray}$

$\begin{eqnarray}
    \frac{\partial}{\partial z}sigmoid(z)=sigmoid(z)\left(1-sigmoid(z)\right) \nonumber
\end{eqnarray}$

```python
import numpy as np
import matplotlib.pyplot as plt

lim = 8
x = np.linspace(-lim, lim, 100)

def  sigmoid(z):
    return 1. / (1. + np.exp(-z))

y = sigmoid(x)

plt.figure(figsize=(6, 4))
plt.xlim(-lim, lim)
plt.ylim(-0.1, 1.1)
plt.title('Sigmoid')
plt.plot(x, y, linestyle='-')
plt.show()
```

```python
y = sigmoid(x) * (1 - sigmoid(x))

plt.figure(figsize=(6, 4))
plt.xlim(-lim, lim)
plt.ylim(-0.01, 0.26)
plt.title('Gradient of Sigmoid')
plt.plot(x, y, linestyle='-')
plt.show()
```
